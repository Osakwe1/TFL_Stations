```python
df_grouped2021=df2021.groupby(['Station']).sum('En/Ex').sort_values('En/Ex', ascending=False).reset_index()
df_grouped2021
```

```python
df_grouped2017=df_grouped2017.drop(columns='NLC')
```

```python
df_merged=pd.merge(
            pd.merge(
                pd.merge(
                    pd.merge(df_grouped2017.drop(columns='NLC'),
                             df_grouped2018.drop(columns='NLC'),
                             on='Station',
                             how='outer'),
                    df_grouped2019.drop(columns='NLC'),
                    on='Station',
                    how='outer'),
                df_grouped2020.drop(columns='NLC'),
                on='Station',
                how='outer'),
            df_grouped2021,
            on='Station',
            how='outer')
df_merged
```

```python
df_merged.rename(columns={"en/ex_x": "En/Ex 2017",
                          "en/ex_y": "En/Ex 2018",
                          "en/ex": "En/Ex 2019", 
                          "En/Ex_x": "En/Ex 2020",
                          "En/Ex_y": "En/Ex 2021"},
                inplace=True)
df_merged
```
